```python
import pandas as pd
import numpy as np

# Реальные средние температуры (примерные данные) для городов по сезонам
seasonal_temperatures = {
    "New York": {"winter": 0, "spring": 10, "summer": 25, "autumn": 15},
    "London": {"winter": 5, "spring": 11, "summer": 18, "autumn": 12},
    "Paris": {"winter": 4, "spring": 12, "summer": 20, "autumn": 13},
    "Tokyo": {"winter": 6, "spring": 15, "summer": 27, "autumn": 18},
    "Moscow": {"winter": -10, "spring": 5, "summer": 18, "autumn": 8},
    "Sydney": {"winter": 12, "spring": 18, "summer": 25, "autumn": 20},
    "Berlin": {"winter": 0, "spring": 10, "summer": 20, "autumn": 11},
    "Beijing": {"winter": -2, "spring": 13, "summer": 27, "autumn": 16},
    "Rio de Janeiro": {"winter": 20, "spring": 25, "summer": 30, "autumn": 25},
    "Dubai": {"winter": 20, "spring": 30, "summer": 40, "autumn": 30},
    "Los Angeles": {"winter": 15, "spring": 18, "summer": 25, "autumn": 20},
    "Singapore": {"winter": 27, "spring": 28, "summer": 28, "autumn": 27},
    "Mumbai": {"winter": 25, "spring": 30, "summer": 35, "autumn": 30},
    "Cairo": {"winter": 15, "spring": 25, "summer": 35, "autumn": 25},
    "Mexico City": {"winter": 12, "spring": 18, "summer": 20, "autumn": 15},
}

# Сопоставление месяцев с сезонами
month_to_season = {
    12: "winter",
    1: "winter",
    2: "winter",
    3: "spring",
    4: "spring",
    5: "spring",
    6: "summer",
    7: "summer",
    8: "summer",
    9: "autumn",
    10: "autumn",
    11: "autumn",
}


# Генерация данных о температуре
def generate_realistic_temperature_data(cities, num_years=10):
    dates = pd.date_range(start="2010-01-01", periods=365 * num_years, freq="D")
    data = []

    for city in cities:
        for date in dates:
            season = month_to_season[date.month]
            mean_temp = seasonal_temperatures[city][season]
            # Добавляем случайное отклонение
            temperature = np.random.normal(loc=mean_temp, scale=5)
            data.append({"city": city, "timestamp": date, "temperature": temperature})

    df = pd.DataFrame(data)
    df["season"] = df["timestamp"].dt.month.map(lambda x: month_to_season[x])
    return df


# Генерация данных
data = generate_realistic_temperature_data(list(seasonal_temperatures.keys()))
data.to_csv("temperature_data.csv", index=False)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from multiprocessing.dummy import Pool as ThreadPool
import time


# Подсчет скользящего среднего окном за 30 дней, выявление аномалий и построение тренда
def analyze_city_data_window(city_data, window=30):
    df = city_data.sort_values("timestamp").copy()
    df["rolling_mean"] = df["temperature"].rolling(window=window).mean()
    df["rolling_std"] = df["temperature"].rolling(window=window).std()

    df["anomaly"] = False
    mask_anomaly = (df["temperature"] > df["rolling_mean"] + 2 * df["rolling_std"]) | (
        df["temperature"] < df["rolling_mean"] - 2 * df["rolling_std"]
    )
    df.loc[mask_anomaly, "anomaly"] = True

    df["day_number"] = (df["timestamp"] - df["timestamp"].min()).dt.days
    coeffs = np.polyfit(df["day_number"], df["temperature"], deg=1)
    df["trend"] = coeffs[0] * df["day_number"] + coeffs[1]

    return df


# Вычисление средней температуры и стандартного отклонения для сезона по городам
def analyze_city_data_avg_and_std(city_data):
    df = city_data.groupby(["city", "season"])["temperature"].agg(["mean", "std"])
    df.reset_index(inplace=True)

    return df


def union_of_two_analysis(df):
    first = analyze_city_data_avg_and_std(df)
    second = analyze_city_data_window(df)
    result = second.merge(first, on=["city", "season"])
    return result


# Последовательный подход
def sequential(df):
    start_time = time.time()

    result_list = []
    for city in df["city"].unique():
        city_data = df[df["city"] == city]
        result = union_of_two_analysis(city_data)
        result_list.append(result)

    full_result = pd.concat(result_list).reset_index(drop=True)
    end_time = time.time()

    print(
        f"Время с использованием последовательного подхода: {end_time - start_time:.2f} с."
    )
    return full_result


# Подход с распараллеливанием
def parallel(df):
    start_time = time.time()
    grouped = [group for _, group in df.groupby("city")]
    with ThreadPool(processes=4) as pool:
        result_list = pool.map(union_of_two_analysis, grouped)

    full_result = pd.concat(result_list).reset_index(drop=True)
    end_time = time.time()

    print(f"Время с использованием распараллеливания: {end_time - start_time:.2f} с.")
    return full_result


# Дополнительные исследования
def correlation_analysis(pivot_df):
    correlation_matrix = pivot_df.corr()
    plt.figure(figsize=(10, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Корреляция температур между городами")
    plt.show()


def compare_seasonal_means(df):
    real_seasonal_means = df.groupby(["city", "season"])["temperature"].mean().unstack()
    print("Средние температуры по сезонам для городов:\n", real_seasonal_means)
```

```python
# Сравнение времени
df = pd.read_csv("temperature_data.csv", parse_dates=["timestamp"])

sequential_result = sequential(df)

parallel_result = parallel(df)
```

```python
# Результаты дополнительного анализа
# Корреляционный анализ
pivot_df = parallel_result.pivot(
    index="timestamp", columns="city", values="temperature"
)
correlation_analysis(pivot_df)

# Сравнение средних по сезонам
compare_seasonal_means(parallel_result)
```

```python
# Получение диапазона температор для текущего дня и средней за все время
def temperature_segment_for_city(df, city: str):
    df["timestamp"] = pd.to_datetime(df["timestamp"])
    df["day_of_year"] = df["timestamp"].dt.dayofyear
    df_berlin = df[df["city"] == city]
    historical_range = (
        df_berlin.groupby("day_of_year")["temperature"]
        .agg(["min", "max"])
        .reset_index()
    )
    historical_avg = (
        df_berlin.groupby("day_of_year")["temperature"].mean().reset_index()
    )
    today_day_of_year = pd.Timestamp.today().day_of_year
    temp_range = historical_range.loc[
        historical_range["day_of_year"] == today_day_of_year, ["min", "max"]
    ]
    avg_temp = historical_avg.loc[
        historical_avg["day_of_year"] == today_day_of_year, "temperature"
    ].values
    return temp_range, avg_temp
```

```python
async def async_approach():
    start_time = time.time()

    async def async_get_temperature_for_city(city):
        async with httpx.AsyncClient() as client:
            geo_url = (
                f"http://api.openweathermap.org/geo/1.0/direct?q={city}&appid={api_key}"
            )
            geo_response = await client.get(geo_url)
            geo_data = geo_response.json()

            lat, lon = geo_data[0].get("lat"), geo_data[0].get("lon")

            weather_url = f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}&units=metric"
            weather_response = await client.get(weather_url)
            weather_data = weather_response.json()

            return weather_data["main"]["temp"]

    async def async_temperature_segment_for_city(df, city: str):
        df["timestamp"] = pd.to_datetime(df["timestamp"])
        df["day_of_year"] = df["timestamp"].dt.dayofyear
        df_berlin = df[df["city"] == city]
        historical_range = (
            df_berlin.groupby("day_of_year")["temperature"]
            .agg(["min", "max"])
            .reset_index()
        )
        historical_avg = (
            df_berlin.groupby("day_of_year")["temperature"].mean().reset_index()
        )
        today_day_of_year = pd.Timestamp.today().day_of_year
        temp_range = historical_range.loc[
            historical_range["day_of_year"] == today_day_of_year, ["min", "max"]
        ]
        avg_temp = historical_avg.loc[
            historical_avg["day_of_year"] == today_day_of_year, "temperature"
        ].values
        return temp_range, avg_temp

    cities = ["Tokyo", "Moscow", "Berlin", "Paris"]

    for city in cities:
        historical_data = await async_temperature_segment_for_city(df, city)
        min_temp, max_temp = (
            historical_data[0]["min"].to_string().split()[1],
            historical_data[0]["max"].to_string().split()[1],
        )
        avg_temp = historical_data[1][0]
        cur_temp = await async_get_temperature_for_city(city)
        if cur_temp is not None:
            print("=" * 80)
            print(f"Текущая температура в городе {city}: {cur_temp}°C")
            if min_temp is not None and max_temp is not None:
                print(
                    f"Диапазон температур за этот день: min: {min_temp}°C, max: {max_temp}°C"
                )
            if avg_temp is not None:
                print(f"Средняя температура за этот день: {avg_temp}°C")
                print(
                    f"Отклонение от средней температуры: {abs(avg_temp - cur_temp)}°C"
                )
            print()

    end_time = time.time()
    print(
        f"Время с использованием асинхронного подхода: {end_time - start_time:.2f} с."
    )
```
